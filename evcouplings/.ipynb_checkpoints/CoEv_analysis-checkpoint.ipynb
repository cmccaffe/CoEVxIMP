# Next we will load in all functions needed to compute the analysis

```python
# define all functions

def distance_calc(tmp_df):

    res_1 = subsubpdb.loc[(subsubpdb['chain_id'] == tmp_df.segment_i) & (subsubpdb['residue_number'] == int(tmp_df.i))]
    res_2 = subsubpdb.loc[(subsubpdb['chain_id'] == tmp_df.segment_j) & (subsubpdb['residue_number'] == int(tmp_df.j))]
    #print(res_1.as_matrix(columns=res_2.columns[11:14]))
    dist = np.linalg.norm(res_1.as_matrix(columns=res_2.columns[11:14])-res_2.as_matrix(columns=res_2.columns[11:14]))
    return dist

def fg(n):
    try:
        a = intra_df.loc[intra_df.MA.ge(n)].index[0]
        return a
    except:
        print('None')
        
def new_coupling_cutoff(tmp_df, cutoff):
    cutoff_threshold = cutoff # the distance cutoff threshold, we will be testing 20, 15, and 10
    top_scores = tmp_df.head(fg(cutoff_threshold))
    cutoff_score = top_scores['cn'].iloc[-1]
    inter_top_scores = top_scores[(top_scores['segment_i']==chain1) & (top_scores['segment_j']==chain2)]
    inter_scores = tmp_df[(tmp_df['segment_i']==chain1) & (tmp_df['segment_j']==chain2)]

    return cutoff_score, inter_top_scores, inter_scores
```

# Function definitions and script pieces

```python
def file_prep(coevfile, pdbfile, chain1, chain2):
    # open coev scores file
    coev = pd.read_csv(coevfile)

    # open pdb as a dataframe
    # extract dataframes with record_name atom and atom_name CB
    #ppdb = PandasPdb().fetch_pdb(pdbfile)
    ppdb = PandasPdb().read_pdb('/home/cns-mccafferty/Mtub_T7SSsecsys/' + pdb_file)


    # extracts all atoms and chains that are represented
    pdbdict = ppdb.df['ATOM']
    subpdb = pdbdict[(pdbdict['atom_name']=='CB')]
    subsubpdb = subpdb[(subpdb['chain_id']==chain1) | (subpdb['chain_id']==chain2)]

    # prepping 
    tmp_df = coev.head(2000)
    tmp_df.loc[tmp_df.segment_i == 'A_1', 'segment_i'] = chain1
    tmp_df.loc[tmp_df.segment_i == 'B_1', 'segment_i'] = chain2
    tmp_df.loc[tmp_df.segment_j == 'A_1', 'segment_j'] = chain1
    tmp_df.loc[tmp_df.segment_j == 'B_1', 'segment_j'] = chain2
    
    return subsubpdb, tmp_df
```

```python
def fg(n):
    try:
        a = intra_df.loc[intra_df.MA.ge(n)].index[0]
        return a
    except:
        print('None')
```

```python
def distance_calc(subsubpdb,tmp_df):

    res_1 = subsubpdb.loc[(subsubpdb['chain_id'] == tmp_df.segment_i) & (subsubpdb['residue_number'] == int(tmp_df.i))]
    res_2 = subsubpdb.loc[(subsubpdb['chain_id'] == tmp_df.segment_j) & (subsubpdb['residue_number'] == int(tmp_df.j))]
    dist = np.linalg.norm(res_1.as_matrix(columns=res_2.columns[11:14])-res_2.as_matrix(columns=res_2.columns[11:14]))
    return dist
```

```python
file_preparation = file_prep('~/Mtub_T7SSsecsys/output/P9WNQ9_P9WNA5/couplings/P9WNQ9_P9WNA5_CouplingScores.csv', 'Mtub_T7SS_all.pdb', 'A', 'B')
```

```python
file_preparation[0]
```

```python
def Main(coevfile, pdbfile, chain1, chain2):
    file_preparation = file_prep(coevfile, pdbfile, chain1, chain2)
    subsubpdb = file_preparation[0]
    tmp_df = file_preparation[1]
    tmp_df['distance'] = tmp_df.apply(lambda subsubpdb: distance_calc(subsubpdb, tmp_df), axis=1)
    top_scores = tmp_df.head(fg(10))
    
    # calculates the rolling mean between intramolecular contacts as you move down score ranks
    intra_df = tmp_df[(tmp_df['segment_i']==tmp_df['segment_j'])]  
    intra_df['MA'] = intra_df['distance'].rolling(10).mean()
    intra_df.to_csv('intra_' +pdbfile+chain1+chain2 + '.csv')
    
    # distances based on the intramolecular contact cutoff
    inter_top_scores = top_scores[(top_scores['segment_i']==chain1) & (top_scores['segment_j']==chain2)] 
    inter_top_scores.to_csv('inter_' +pdbfile+chain1+chain2 + '.csv')
```

```python
def distance_calc(tmp_df):

    res_1 = pdbdict.loc[(pdbdict['chain_id'] == tmp_df.segment_i) & (pdbdict['residue_number'] == int(tmp_df.i))]
    res_2 = pdbdict.loc[(pdbdict['chain_id'] == tmp_df.segment_j) & (pdbdict['residue_number'] == int(tmp_df.j))]
    #print(res_1.as_matrix(columns=res_2.columns[11:14]))
    if res_1.empty:
        print('DataFrame is empty!')
    if res_2.empty:
        print('DataFrame is empty!')
    else:
        #print(res_1.as_matrix(columns=res_2.columns[11:14]).shape)
        #print(res_2.as_matrix(columns=res_2.columns[11:14]).shape)
        #print(res_2)
        dist = np.linalg.norm(res_1.as_matrix(columns=res_2.columns[11:14])-res_2.as_matrix(columns=res_2.columns[11:14]))
        print(dist)
        return dist
```

```python
frame10cutoff['model_distance'] = frame10cutoff.apply(distance_calc, axis=1)
```
